```python
import pandas as pd
import numpy as np
```

```python
data =  pd.read_csv('Sample data_IMDB-Movie-Data.csv')
```

```python
data.shape
```

```python
data = data.drop('Unnamed: 12', axis=1)
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
d = data['Genre'].str.split(',')
```

```python
ls = [x for sub in d for x in sub ]
```

```python
genres = set(ls)
```

```python
genres
```

```python
data = data.drop(['Title', 'Actors', 'Director'], axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(vocabulary=list(genres), lowercase=False)
```

```python
 x= vectorizer.fit_transform(data['Genre'])
```

```python
x
```

```python
col = vectorizer.get_feature_names_out()
```

```python
dx = list(col)
```

```python
df = pd.DataFrame(x.toarray(), columns=dx)
```

```python
df
```

```python
data = pd.concat([data,df], axis=1)
```

```python
data.head()
```

```python
data = data.drop('Genre', axis=1)
```

```python
data.to_csv('processed.csv', index=False)
```
